### Importing Relevant Packages

```python
import numpy as np 
from matplotlib import pyplot as plt 
from osgeo import gdal 
from scipy.interpolate import RegularGridInterpolator
import rasterio 
from rasterio.warp import calculate_default_transform, reproject, Resampling
import matplotlib.gridspec as gridspec
import csv
import pandas as pd
```

### Creating Functions to Analyse the Data

```python
# This function is used for Figure 1 and 2. It allows us to extract x, y, and z values from a GeoTIFF image.

def xyz_geotiff(img):
    band = img.GetRasterBand(1)
    z = np.flipud(band.ReadAsArray()).T 
    width = img.RasterXSize 
    height = img.RasterYSize 
    gt = img.GetGeoTransform() 
    minx = gt[0] 
    miny = gt[3] + width*gt[4] + height*gt[5] 
    maxx = gt[0] + width*gt[1] + height*gt[2] 
    maxy = gt[3] 
    lon = np.linspace(minx,maxx,width) 
    lat = np.linspace(miny,maxy,height) 
    return lon,lat,z 

# This function is used for Figure 1. It interpolates along a 2D transect.

def transect2dxyz(a0,a1,x,y,z): 
    tempinterpolator = RegularGridInterpolator((x,y),z, bounds_error=False) 
    linelength = ((a1[0]-a0[0])**2+(a1[1]-a0[1])**2)**0.5 
    spoints = np.arange(0,linelength,np.diff(x)[0]) 
    xpoints = np.cos(np.arctan2(a1[1]-a0[1],a1[0]-a0[0]))*spoints+a0[0] 
    ypoints = np.sin(np.arctan2(a1[1]-a0[1],a1[0]-a0[0]))*spoints+a0[1]  
    zpoints = tempinterpolator((xpoints,ypoints)) 
    return spoints,xpoints,ypoints,zpoints 

# This function is used for Figure 2. It calculates the R-Squared value.

def get_r2(x,y): 
    polyfit = np.polyfit(x,y,1)
    y_predicted = np.polyval(polyfit,x)
    residuals = y-y_predicted
    return 1-np.sum(residuals**2)/np.sum((y-np.mean(y))**2)
```

### Changing the Coordinate System

```python
# The raw NetCDF files downloaded from open source websiites (NSIDC) were imported into QGIS and then extracted as raw GeoTIFF files.
# The GeoTIFF files were downloaded in polar stereogrpahic EPSG:3031 (northings and eastings), and need to be reprojected into EPSG:4326 if latitude and longitude coordinates are to be used.

# The following code is applicable to Figure 1.

# To reproject the Cook Ice Shelf Surface Elevation profile:

dst_crs = 'EPSG:4326'

with rasterio.open('Fig1BedMachineAntarcticav3SurfaceGeometry.tif') as src: 
    transform, width, height = calculate_default_transform(
        src.crs, dst_crs, src.width, src.height, *src.bounds)
    kwargs = src.meta.copy()
    kwargs.update({
        'crs': dst_crs,
        'transform': transform,
        'width': width,
        'height': height
    })

    with rasterio.open('BedMachSurf4326Fig1A.tif', 'w', **kwargs) as dst: 
        for i in range(1, src.count + 1):
            reproject(
                source=rasterio.band(src, i),
                destination=rasterio.band(dst, i),
                src_transform=src.transform,
                src_crs=src.crs,
                dst_transform=transform,
                dst_crs=dst_crs,
                resampling=Resampling.bilinear)

# To reproject the Cook Ice Shelf Bed Elevation profile:

with rasterio.open('Fig1BedMachineAntarcticav3BedGeometry.tif') as src: 
    transform, width, height = calculate_default_transform(
        src.crs, dst_crs, src.width, src.height, *src.bounds)
    kwargs = src.meta.copy()
    kwargs.update({
        'crs': dst_crs,
        'transform': transform,
        'width': width,
        'height': height
    })

    with rasterio.open('BedMachBed4326Fig1C.tif', 'w', **kwargs) as dst: 
        for i in range(1, src.count + 1):
            reproject(
                source=rasterio.band(src, i),
                destination=rasterio.band(dst, i),
                src_transform=src.transform,
                src_crs=src.crs,
                dst_transform=transform,
                dst_crs=dst_crs,
                resampling=Resampling.bilinear)
```

### Creating Figure 1

```python
a0 = [152.9914245605469887, -68.1560134887695028]
a1 = [152.5248260498050001, -68.9364395141601989] 

data_geotiff = gdal.Open('BedMachSurf4326Fig1A.tif') 
lon_geotiff,lat_geotiff,z_geotiff = xyz_geotiff(data_geotiff) 
spoints,lonpoints,latpoints,zpoints = transect2dxyz(a0,a1,lon_geotiff,lat_geotiff,z_geotiff) 
z_geotiff[z_geotiff>1e+30] = np.nan 

figure1 = plt.gcf() 
plt.figure().set_size_inches(180/25.4,110/25.4,forward=True)
gs = gridspec.GridSpec(6, 2, height_ratios=[1, 1, 1, 1, 1, 1])
plt.rcParams.update({'font.size': 6})  
plt.rcParams['font.family'] = 'Arial'

subplot1 = plt.subplot(gs[0:3, 0]) 
plt.pcolormesh(lon_geotiff,lat_geotiff,z_geotiff.T,cmap='magma') 
plt.colorbar(label='Elevation (m)') 
subplot1.tick_params(axis='both', direction='in')
subplot1.plot(lonpoints,latpoints,c='y') 
subplot1.set_xlabel('Longitude ($\\degree$)') 
subplot1.set_ylabel('Latitude ($\\degree$)') 

subplot2 = plt.subplot(gs[0:2, 1]) 
subplot2.plot(spoints,zpoints,c='y') 
subplot2.tick_params(axis='both', direction='in') 
subplot2.set_xlabel('Distance Along Transect ($\\degree$)') 
subplot2.set_ylabel('Elevation (m)') 
subplot2.set_xlim(spoints[0],spoints[-1]) 


data_geotiff2 = gdal.Open('BedMachBed4326Fig1C.tif') 
lon_geotiff2,lat_geotiff2,z_geotiff2 = xyz_geotiff(data_geotiff2) 
spoints2,lonpoints2,latpoints2,zbedpoints2 = transect2dxyz(a0,a1,lon_geotiff2,lat_geotiff2,z_geotiff2) 
z_geotiff2[z_geotiff2<-8000] = np.nan 

subplot3 = plt.subplot(gs[3:6, 0]) 
plt.pcolormesh(lon_geotiff2,lat_geotiff2,z_geotiff2.T,cmap='magma') 
plt.colorbar(label='Elevation (m)') 
subplot3.tick_params(axis='both', direction='in') 
subplot3.set_xlabel('Longitude ($\\degree$)') 
subplot3.set_ylabel('Latitude ($\\degree$)') 
subplot3.plot(lonpoints2,latpoints2,c='y') 
subplot3.set_xlabel('Longitude ($\\degree$)') 
subplot3.set_ylabel('Latitude ($\\degree$)') 

subplot4 = plt.subplot(gs[2:4,1]) 
subplot4.plot(spoints2,zbedpoints2,c='y') 
subplot4.tick_params(axis='both', direction='in') 
subplot4.set_xlabel('Distance Along Transect ($\\degree$)') 
subplot4.set_ylabel('Elevation (m)') 

subplot5 = plt.subplot(gs[4:6,1])
subplot5.plot(spoints,zpoints,c='c', label='Surface Elevation')
subplot5.plot(spoints,zbedpoints2,c='m', label='Bed Elevation')
subplot5.set_xlabel('Distance Along Transect ($\\degree$)')
subplot5.set_ylabel('Elevation / Thickness (m)')
zbedthick = zpoints - zbedpoints2
subplot5.plot(spoints,zbedthick,c='g', label='Ice Thickness', linestyle='--')
subplot5.set_ylim(-850,1500)
subplot5.legend(frameon=False,loc='upper left', fontsize=5, ncol=2)
subplot5.tick_params(axis='both', direction='in')

subplot1.text(-0.1, 1.0, 'a', transform=subplot1.transAxes, fontsize=8, fontweight='bold')
subplot2.text(-0.1, 1.0, 'b', transform=subplot2.transAxes, fontsize=8, fontweight='bold')
subplot3.text(-0.1, 1.0, 'c', transform=subplot3.transAxes, fontsize=8, fontweight='bold')
subplot4.text(-0.1, 1.0, 'd', transform=subplot4.transAxes, fontsize=8, fontweight='bold')
subplot5.text(-0.1, 1.10, 'e', transform=subplot5.transAxes, fontsize=8, fontweight='bold')

plt.tight_layout()

plt.savefig('Figure1.png', dpi=500, bbox_inches='tight')
```

### Extracting Data from GeoTIFF Files (For Figure 2)

```python
# For Thickness Data:

# EXTRACTS THE DESIGNATED BAND FOR THE SPECIIC YEAR 

Fig2AThicknessResults = [("Band", "Mean Anomaly", "Median Anomaly")]

for band_number in range(1, 105):
    raster_path = "Fig2QuarterlyThickness.tif"
    with rasterio.open(raster_path) as src:
        band_data = src.read(band_number) 
        
    out_raster = f"Fig2Aband{band_number}.tif"
    with rasterio.open(raster_path) as src:
        profile = src.profile  
        profile.update(count=1) 
    
        with rasterio.open(out_raster, "w", **profile) as dst:
            dst.write(band_data, 1)
    
    ##########################################
    
    # CALCULATES THE ANOMALY BETWEEN THE SPECIFIC YEAR BAND AND THE MEAN BAND
    
    raster1_path = f"Fig2Aband{band_number}.tif"  
    raster2_path = "Fig2MeanThickness.tif" 
    output_path = f"Fig2Aanom{band_number}.tif"
    
    with rasterio.open(raster1_path) as src1, rasterio.open(raster2_path) as src2:
        raster1 = src1.read(1) 
        raster2 = src2.read(1)  
    
        difference = raster1 - raster2
    
        difference[difference == 0] = np.nan 
    
        profile = src1.profile
        profile.update(dtype=rasterio.float32, count=1, nodata=np.nan)
    
        with rasterio.open(output_path, "w", **profile) as dst:
            dst.write(difference, 1)
        
    ##########################################
    
    # CALCULATE THE MEAN AND MEDIAN OF THE ANOMALY ACROSS THE ENTIRE ICE SHEET
    
    raster3_path = f"Fig2Aanom{band_number}.tif"
    
    with rasterio.open(raster3_path) as src:
        
        raster3_data = src.read(1)
        
        nodata_value = src.nodata
    
        if nodata_value is not None:
            raster3_data = np.where(raster3_data == nodata_value, np.nan, raster3_data)
    
        mean_value = np.nanmean(raster3_data)
        median_value = np.nanmedian(raster3_data)

    Fig2AThicknessResults.append((band_number, mean_value, median_value))

output_csv = "Fig1AThicknessResults.csv"

with open(output_csv, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(Fig2AThicknessResults)
```

```python
# For Melt Rate Data:

# EXTRACTS THE DESIGNATED BAND FOR THE SPECIIC YEAR

Fig2BMeltResults = [("Band", "Mean Anomaly", "Median Anomaly")]

for band_number in range(1, 105):
    raster_path = "Fig2QuarterlyMeltRate.tif"
    with rasterio.open(raster_path) as src:
        band_data = src.read(band_number) 
        
    out_raster = f"Fig2Bband{band_number}.tif"
    with rasterio.open(raster_path) as src:
        profile = src.profile  
        profile.update(count=1) 
    
        with rasterio.open(out_raster, "w", **profile) as dst:
            dst.write(band_data, 1)
    
    ##########################################
    
    # CALCULATES THE ANOMALY BETWEEN THE SPECIFIC YEAR BAND AND THE MEAN BAND
    
    raster1_path = f"Fig2Bband{band_number}.tif"  
    raster2_path = "Fig2MeanMeltRate.tif" 
    output_path = f"Fig2Banom{band_number}.tif"
    
    with rasterio.open(raster1_path) as src1, rasterio.open(raster2_path) as src2:
        raster1 = src1.read(1) 
        raster2 = src2.read(1)  
    
        difference = raster1 - raster2
    
        difference[difference == 0] = np.nan 
    
        profile = src1.profile
        profile.update(dtype=rasterio.float32, count=1, nodata=np.nan)
    
        with rasterio.open(output_path, "w", **profile) as dst:
            dst.write(difference, 1)
        
    ##########################################
    
    # CALCULATE THE MEAN AND MEDIAN OF THE ANOMALY ACROSS THE ENTIRE ICE SHEET
    
    raster3_path = f"Fig2Banom{band_number}.tif"
    
    with rasterio.open(raster3_path) as src:
        
        raster3_data = src.read(1)
        
        nodata_value = src.nodata
    
        if nodata_value is not None:
            raster3_data = np.where(raster3_data == nodata_value, np.nan, raster3_data)
    
        mean_value = np.nanmean(raster3_data)
        median_value = np.nanmedian(raster3_data)

    Fig2BMeltResults.append((band_number, mean_value, median_value))

output_csv = "Fig2BMeltResults.csv"

with open(output_csv, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(Fig2BMeltResults)
```

### Formatting the CSV Tables to Prepare for Plotting

```python
# The ' Fig2Time ' file lists the band number and their respective band information (days since 1/1/1990).
# I had to manually extract the band information from QGIS (when downloading the raster files) and write it in - there was no way to automate this step.

######################################################################################
# See the attached word doc file ' Fig2Time Explanation ' file in GitHUB for more info
######################################################################################

Fig2Time = pd.read_csv('Fig2TimeInputs.csv')

Fig1AThick = pd.read_csv('Fig1AThicknessResults.csv')
Fig1AThick['Year'] = (Fig2Time['BandInfo:DaysSince1/1/1950'] / 365) + 1950
Fig1AThick.to_csv('Fig1AThickTime.csv', index=False)
Fig1AThickTime = pd.read_csv('Fig1AThickTime.csv')

Fig1BMelt = pd.read_csv('Fig2BMeltResults.csv')
Fig1BMelt['Year'] = (Fig2Time['BandInfo:DaysSince1/1/1950'] / 365) + 1950
Fig1BMelt.to_csv('Fig1BMeltTime.csv', index=False)
Fig1BMeltTime = pd.read_csv('Fig1BMeltTime.csv')
```

### Plotting Figure 2

```python
# Defining and x and y coordinates

x1 = Fig1AThickTime['Year']
y1 = Fig1AThickTime['Mean Anomaly']
y2 = Fig1AThickTime['Median Anomaly']
xA = Fig1BMeltTime['Year']
yA = Fig1BMeltTime['Mean Anomaly']
yB = Fig1BMeltTime['Median Anomaly']

# Defining the linear regression for the Mean Anomalies

Fig2Alinearpolyfit = np.polyfit(x1,y1,1)
m1, c1 = Fig2Alinearpolyfit
y_fit = m1 * x1 + c1
Fig2Blinearpolyfit = np.polyfit(xA,yA,1)
m2, c2 = Fig2Blinearpolyfit
y_fit = m2 * xA + c2

# Calculating the R-Squared valus:

ThickRSquared = get_r2(x1,y1)
MeltRSquared = get_r2(xA,yB)
```

```python
# Plotting Figure 2:

plt.figure().set_size_inches(180/25.4,90/25.4,forward=True)
plt.rcParams.update({'font.size': 6})
plt.rcParams['font.family'] = 'Arial'

subplot1 = plt.subplot(211)
subplot1.scatter(x1,y1,label='Mean Thickness Anomaly', marker='x', s=10, linewidths=0.5, c='m')
subplot1.scatter(x1,y2,label='Median Thickness Anomaly', marker='x', s=10, linewidths=0.5, c='c')
subplot1.set_xlabel('Time (Year)') 
subplot1.set_ylabel('Thickness Anomaly (m)') 
subplot1.plot(x1,np.polyval(Fig2Alinearpolyfit, x1), c='r', alpha=0.8, lw=1, label = 'Mean: y = %.2fx %.2f\nMean: R$^2=%.2f$' % (m1, c1, ThickRSquared) )
subplot1.legend(frameon=False,loc='upper left')
subplot1.tick_params(axis='both', direction='in')

subplot2 = plt.subplot(212)
subplot2.scatter(xA,yA,label='Mean BMR Anomaly', marker='x', s=10, linewidths=0.5, c='m')
subplot2.scatter(xA,yB,label='Median BMR Anomaly', marker='x', s=10, linewidths=0.5, c='c')
subplot2.set_xlabel('Time (Year)') 
subplot2.set_ylabel('BMR Anomaly (m)') 
subplot2.plot(x1,np.polyval(Fig2Blinearpolyfit, xA), c='r', alpha=0.8, lw=1, label = 'Mean: y = %.2fx + %.2f\nMean: R$^2=%.2f$' % (m2, c2, MeltRSquared) )
subplot2.legend(frameon=False)
subplot2.tick_params(axis='both', direction='in')

subplot1.text(-0.08, 1.0, 'a', transform=subplot1.transAxes, fontsize=8, fontweight='bold')
subplot2.text(-0.08, 1.0, 'b', transform=subplot2.transAxes, fontsize=8, fontweight='bold')

plt.subplots_adjust(hspace=0.3)

plt.savefig('Figure2.png', dpi=500, bbox_inches='tight')
```

### Other Calculations

```python
raster1992_path = f"Fig2Aband1.tif"
with rasterio.open(raster1992_path) as src:
        raster1992_path = src.read(1)
        nodata_value = src.nodata
        if nodata_value is not None:
            raster1992_path = np.where(raster1992_path == nodata_value, np.nan, raster1992_path)
        mean_1992Thickness = np.nanmean(raster1992_path)

raster2017_path = f"Fig2Aband104.tif"
with rasterio.open(raster2017_path) as src:
        raster2017_path = src.read(1)
        nodata_value = src.nodata
        if nodata_value is not None:
            raster2017_path = np.where(raster2017_path == nodata_value, np.nan, raster2017_path)
        mean_2017Thickness = np.nanmean(raster2017_path)

ThicknessChangeHeightMetres = mean_2017Thickness - mean_1992Thickness
ThicknessChangeVolumeCubicMetres = ThicknessChangeHeightMetres * 3.5e+9 # Cook Ice Shelf is ~ 3,500 km^2 = 3.5e+9 m^2 (Lee et al. 2023)
ThicknessChangeVolumeCubicKilometres = ThicknessChangeVolumeCubicMetres / 1e+9
GlobalSeaLevelChange = 2.5 / 1000 * ThicknessChangeVolumeCubicKilometres # Based on the estimate of 2.5mm global sea level change is caused by 1000 km^3 of ice (Fairbridge and Jelgersma 1990).

print(f"The thickness of the Cook Ice Shelf in 1992 is {mean_1992Thickness:.2f} metres.")
print(f"The thickness of the Cook Ice Shelf in 2017 is {mean_2017Thickness:.2f} metres.")
print(f"This means that the change in height of this ice shelf is ~ {ThicknessChangeHeightMetres:.2f} metres.")
print(f"This means that the change in volume of this ice shelf is ~ {ThicknessChangeVolumeCubicMetres:.2f} cubic metres, or {ThicknessChangeVolumeCubicKilometres} cubic kilometres.")
print(f"This means that the change in global sea level due to thickening of the Cook Ice Shelf between 1992 and 2017 is equivalent to -{GlobalSeaLevelChange:.2f} mm.")

```
